```python
plt.boxplot(df.price)

Q1 = df.price.quantile(0.25)  
Q3 = df.price.quantile(0.75)  
IQR = Q3 - Q1

df = df[(df.price >= Q1 - 1.5 * IQR) & (df.price <= Q3 + 1.5 * IQR)]
```

```python
plt.boxplot(df.area)

Q1 = df.area.quantile(0.25)  
Q3 = df.area.quantile(0.75)  
IQR = Q3 - Q1
df = df[(df.area >= Q1 - 1.5 * IQR) & (df.area <= Q3 + 1.5 * IQR)]
```

```python
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

def binary_map(x):
    return x.map({'yes': 1, 'no': 0})

df[varlist] = df[varlist].apply(binary_map)
```

```python
df.head()
```

```python
furnishing_dummies = pd.get_dummies(df['furnishingstatus'], drop_first=True)
df = pd.concat([df, furnishing_dummies], axis=1)
df.drop('furnishingstatus', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['semi-furnished'] = df['semi-furnished'].astype(int)
```

```python
df.drop('unfurnished',axis = 1, inplace=True)
```

```python
df.head()
```

```python
scaler = MinMaxScaler()
num_var = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']
df[num_var] = scaler.fit_transform(df[num_var])
```

```python
x = df.drop('price', axis = 1)
y = df['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=24)
```

```python
df.head()
```

```python
y_train.head()
```

```python
lm = LinearRegression()
```

```python
model = lm.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```
